```python
import pandas as pd
import geopandas as gpd
import json
```

# first

```python
bldg_path = './00_final_bldgs_key.geojson'
biz_path = './01_biz_bldgs.geojson'
```

```python
bldg = gpd.read_file(bldg_path)
```

```python
target_crs = 'EPSG:4326'
bldg = bldg.to_crs(target_crs)
```

```python
bldg['center'] = bldg.geometry.centroid
```

```python
bldg['center_x'] = bldg['center'].x
bldg['center_y'] = bldg['center'].y
```

```python
bldg = bldg.drop(columns='center')
```

```python
bldg_df = bldg.drop(columns='geometry')
```

```python
bldg_df.head(10)
```

```python
bldg_df.to_csv('../bldg.csv',index=False)
```

#biz

```python
biz = gpd.read_file(biz_path)
```

```python
target_crs = 'EPSG:4326'
biz = biz.to_crs(target_crs)
```

```python
biz['latitude'] = biz.geometry.y
biz['longitude'] = biz.geometry.x
```

```python
biz_df = biz.drop(columns='geometry')
```

```python
biz_df.head(10)
```

```python
biz_df.to_csv('../biz.csv',index=False)
```

```python
import geojson
```

```python
dist_list = pd.read_csv('../../../data/1_dist.csv')
```

```python
dist_list.head(2)
```

```python
features = []
```

```python
for index, row in dist_list.iterrows():
    point = geojson.Point((row['longitude'], row['latitude']))
    properties  = {
        'Distributor Name':row['CUST__NAME'],
        'Supplier Specialty': row['supply_desc']
    }
    features.append(geojson.Feature(geometry=point, properties=properties))

feature_collection = geojson.FeatureCollection(features)
```

```python
geojson_string = geojson.dumps(feature_collection, indent=2)
```
